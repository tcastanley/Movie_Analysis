# Data Cleaning

## Cleaning the file names

```python
csv_files_dict = {}
for filename in csv_files:
    filename_cleaned = os.path.basename(filename).replace(".csv", "").replace(".", "_").replace("_gz", "") # cleaning the filenames
    filename_df = pd.read_csv(filename, index_col=0)
    csv_files_dict[filename_cleaned] = filename_df
```

# Creating new database in SQL

```python
import sqlite3
```

## Making SQL file

```python
conn = sqlite3.connect("movies_db.sqlite")
cur = conn.cursor()
```

## Creating function to create SQL table from dataframe

```python
def create_sql_table_from_df(df, name, conn):
    try:
        df.to_sql(name, conn, if_exists='fail')
        print(f"Created table {name}")
    
    except Exception as e:
        print(f"could not make table {name}")
        print(e)
```

## For loop to iterate through collection of CSV items

```python
for name, table in csv_files_dict.items():
    create_sql_table_from_df(table, name, conn)
```

```python
conn.execute("select name from sqlite_master where type='table';").fetchall()
```
